**Exercice 1**

```python
import matplotlib.pyplot as plt
import numpy as np

# Définition de l'équation de la courbe
def equation(x):
    return x**5
# Trouver des valeurs pour x dans l'intervalle [0, 1]
valeur_x = np.linspace(0, 1, 1000)
# Trouver les valeurs de y en utilisant l'équation
valeur_y = equation(valeur_x)

# La courbe d'équation x⁵
plt.plot(valeur_x, valeur_y, label='y = x⁵',color='red')

# Le carré délimité par les droites x=0, x=1, y=0, y=1
carre_x = [0, 1, 1, 0, 0]
carre_y = [0, 0, 1, 1, 0]
plt.plot(carre_x, carre_y, label='Carré', linestyle='-', color='#8B0000')

# Etiquettes et une légende
plt.xlabel('x')
plt.ylabel('y')
plt.suptitle('Figure 1: Courbe d\'équation x⁵ sur l’intervalle [0;1]')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=3)
# Affichage du graphique
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Creation de la classe Curve
class Curve:
    def __init__(self, start=0, stop=1, nbr_points=5432):
        self.start = start
        self.stop = stop
        self.nbr_points = nbr_points

    def _equation(self, x):
        return x ** 5

    def graphique_curve(self):
        # Trouver des valeurs pour x dans l'intervalle [start, stop]
        valeur_x = np.linspace(self.start, self.stop, self.nbr_points)

        # Trouver les valeurs de y en utilisant l'équation
        y_values = self._equation(valeur_x)
        # La courbe d'équation x⁵
        plt.plot(valeur_x, y_values, label='y = x⁵', color='red')

        # Le carré délimité par les droites x=start, x=stop, y=0, y=max(y_values)
        carre_x = [self.start, self.stop, self.stop, self.start, self.start]
        carre_y = [0, 0, max(y_values), max(y_values), 0]
        plt.plot(carre_x, carre_y, label='Carré', linestyle='-', color='#8B0000')

        # Génération des points
        points = np.column_stack((valeur_x, np.random.uniform(0, max(valeur_y), len(valeur_x))))

        # Séparation des points au-dessus et en dessous de la courbe
        point_dessus = [point for point in points if point[1] > self._equation(point[0])]
        point_dessous = [point for point in points if point[1] <= self._equation(point[0])]

        x_dessus, y_dessus = zip(*point_dessus)
        x_dessous, y_dessous = zip(*point_dessous)

        # Affichage des points au-dessus de la courbe en bleu et en dessous en vert
        plt.plot(x_dessus, y_dessus, 'bx', label='Points au dessus de la courbe')
        plt.plot(x_dessous, y_dessous, 'go', label='Points en dessous de la courbe')
        # Etiquettes et légende
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=3)
        # Affichage du graphique
Curve().graphique_curve()
plt.suptitle('Figure 1: Courbe d\'équation x⁵ sur l’intervalle [0;1] avec les points')
```

**Calcul des surface des deux types de points**

*Pour calculer la surface en bleu et celle en vert, nous pouvez utiliser la méthode des trapèzes pour approximer les intégrales sous les courbes bleues et vertes. C'est ce que nous avons utilise pour le calcul.*

```python
import matplotlib.pyplot as plt
import numpy as np

# Création de la classe Curve
class Curve:
    def __init__(self, start=0, stop=1, nbr_points=5432):
        self.start = start
        self.stop = stop
        self.nbr_points = nbr_points

    def _equation(self, x):
        return x ** 5

    def graphique_curve(self):
        # Trouver des valeurs pour x dans l'intervalle [start, stop]
        valeur_x = np.linspace(self.start, self.stop, self.nbr_points)

        # Trouver les valeurs de y en utilisant l'équation
        y_values = self._equation(valeur_x)

        # La courbe d'équation x⁵
        plt.plot(valeur_x, y_values, label='y = x⁵', color='red')

        # Le carré délimité par les droites x=start, x=stop, y=0, y=max(y_values)
        carre_x = [self.start, self.stop, self.stop, self.start, self.start]
        carre_y = [0, 0, max(y_values), max(y_values), 0]
        plt.plot(carre_x, carre_y, label='Carré', linestyle='-', color='#8B0000')

        # Génération des points
        points = np.column_stack((valeur_x, np.random.uniform(0, max(y_values), len(valeur_x))))

        # Séparation des points au-dessus et en dessous de la courbe
        point_dessus = [point for point in points if point[1] > self._equation(point[0])]
        point_dessous = [point for point in points if point[1] <= self._equation(point[0])]

        x_dessus, y_dessus = zip(*point_dessus)
        x_dessous, y_dessous = zip(*point_dessous)

        # Affichage des points au-dessus de la courbe en bleu et en dessous en vert
        plt.plot(x_dessus, y_dessus, 'bx', label='Points au dessus de la courbe')
        plt.plot(x_dessous, y_dessous, 'go', label='Points en dessous de la courbe')

        # Calcul des surfaces en bleu et en vert
        surface_bleu = np.trapz(y_dessus, x_dessus)
        surface_vert = np.trapz(y_dessous, x_dessous)
        # Affichage des surfaces calculées
        print(f"Surface en bleu est egale a: {surface_bleu}")
        print(f"Surface en vert est egale a: {surface_vert}")

        # Etiquettes et légende
        plt.xlabel('x')
        plt.ylabel('y')
        plt.suptitle('Figure 2: Courbe d\'équation x⁵ sur l’intervalle [0;1] avec les points')
        plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=3)
        # Affichage du graphique
Curve().graphique_curve()
```
